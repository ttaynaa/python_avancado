6) Na base de dados, qual a porcentagem de países presentes na União Europeia?

```python
valores = data['European Union Membership'].value_counts()

print('Não membros:', round(valores[0]/len(data),2))
print('Membros:', round(valores[1]/len(data), 2))
```

7) Crie uma nova classificação de índice baixo, médio, alto e muito alto, para o empreendedoorismo feminino em cada um países. Para isso tome como intervalo os valores dos quartis dessa *feature*. Em qual classficação se encontra o Brasil? Quantos foram os países em cada uma das categorias?.

```python
bins = data['Women Entrepreneurship Index'].describe()
bins = bins[3:,].values
```

```python
data['classificacao'] = pd.cut(data['Women Entrepreneurship Index'],bins=bins, 
                               labels=["baixo", "medio", "alto",'muito alto'],
                               include_lowest=True)

print(data.classificacao.value_counts())
data[data.Country == 'Brazil']
```
