Antes de começar a próxima questão, precisamos rever alguns parâmetros importantes do Pairplot:

* **hue**: variável no dataset para separar em diferentes grupos de acordo com as classes presentes nesses dados
* **kind**: tipo de gráfico a se construir ('scatter', 'kde', 'hist', 'reg')
* **diag_kind**: tipo de gráfico a ser construido na driagonal principal ('auto', 'hist', 'kde', None)
* **vars (x_vars, y_vars)**: parâmetro para selecionar as variáveis a serem usadas para gerar o gráfico.

Para ver os demais veja esse [link](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

3) Crie um Pairplot com os dados dos pinguins com as seguintes especificações:

* Pairplot simples
* Pairplot com o parâmetro *hue* para espécies, nome da ilha ou sexo dos pinguins
* Pairplot com diferentes variáveis para o parâmetro *kind*

Quais análises podemos tirar dos gráficos feitos?

```python
sns.pairplot(penguins)
```

```python
#Nessa célular variar o hue para sexo, species e island
sns.pairplot(penguins, hue="island")
```

```python
sns.pairplot(penguins, kind="hist", diag_kind='kde')
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```

4) Com a ajuda dos parâmetros *x_vars* e *y_vars*, crie um pairplot apenas com as variáveis "bill_length_mm", "bill_depth_mm", "flipper_length_mm" no eixo x e apenas duas delas no eixo y. Faça uma classificação a partir das ilhas.

```python
sns.pairplot(
    penguins,
    x_vars=["bill_length_mm", "bill_depth_mm", "flipper_length_mm"],
    y_vars=["bill_length_mm", "bill_depth_mm"], hue='island',
    palette= "prism"
)
```

5) Crie um *Jointplot* com as colunas *bill_length_mm* e *bill_depth_mm* e atribua a uma variável *g*. Em seguida, a partir dessa variável chame os métodos *plot_join* e *plot_marginals* [veja a documentação aqui](http://seaborn.pydata.org/generated/seaborn.JointGrid.html#seaborn.JointGrid) para melhorar o aspecto do gráfico. Além disso, separe os pinguins pela categoria *species* com a ajuda do parâmetro *hue*.

```python
g = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue='species', height = 10)

g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", clip_on=False)
```
