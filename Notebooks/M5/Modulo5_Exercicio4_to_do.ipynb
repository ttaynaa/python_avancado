Durante esse exercício vamos trabalhar com uma base de dados nativa do **Seaborn**: Palmer Archipelago (Antarctica) penguin data. 

Os dados foram coletados e disponibilizados pela Dra. Kristen Gorman e pela Palmer Station, Antarctica LTER, membro da Long Term Ecological Research Network. [GitHub oficial](https://github.com/allisonhorst/palmerpenguins/blob/master/README.md)

* **species**: espécie do pinguim  (Chinstrap, Adélie, or Gentoo)

Antes de começar a próxima questão, precisamos rever alguns parâmetros importantes do Pairplot:

* **hue**: variável no dataset para separar em diferentes grupos de acordo com as classes presentes nesses dados
* **kind**: tipo de gráfico a se construir ('scatter', 'kde', 'hist', 'reg')
* **diag_kind**: tipo de gráfico a ser construido na driagonal principal ('auto', 'hist', 'kde', None)
* **vars (x_vars, y_vars)**: parâmetro para selecionar as variáveis a serem usadas para gerar o gráfico.

Para ver os demais veja esse [link](https://seaborn.pydata.org/generated/seaborn.pairplot.html)
